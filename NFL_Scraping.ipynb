```python
nfl_years_path = Path("/Users/camsmithers/Desktop/Camalytics/CamalyticsEnv/Projects/Sports/NFL/Data/Season")
nfl_game_list = []
```

```python
for filename in os.listdir(nfl_years_path):
    if filename.endswith(".html"):
        with open(os.path.join(nfl_years_path, filename), "r", encoding="utf-8") as f:
            soup = BeautifulSoup(f, "html.parser")
            links = [a["href"] for a in soup.find_all("a", href=True) if "boxscores/20" in a["href"]]

            for link in links:
                full_link = f"https://www.pro-football-reference.com{link}"
                nfl_game_list.append(full_link)
```

```python
def nfl_chunk_list(games_listed, number_of):
    for i in range(0, len(games_listed), number_of):
        yield games_listed[i:i + number_of]
```

```python
nfl_chunk_1, nfl_chunk_2, nfl_chunk_3 = list(nfl_chunk_list(nfl_game_list, 400))
```

```python
nfl_game_folder = "/Users/camsmithers/Desktop/Camalytics/CamalyticsEnv/Projects/Sports/NFL/Data/NFLGames/NFL_3/"

for game in nfl_chunk_3:
    url = game
    data = requests.get(game)

    file_path = nfl_game_folder + game.split("/")[-1] + "l"

    with open(file_path, "w+", encoding="utf-8") as f:
        f.write(data.text)
    time.sleep(random.uniform(5, 10))
```

```python
nfl_afc_abb = ["buf", "mia", "nyj", "nwe", "rav", "pit", "cin", "cle", "htx", "clt", "jax", "oti", "kan", "sdg", "den", "rai"]
nfl_nfc_abb = ["phi", "was", "dal", "nyg", "det", "min", "gnb", "chi", "tam", "atl", "car", "nor", "ram", "sea", "crd", "sfo"]
```

```python
nfl_team_folder = "/Users/camsmithers/Desktop/Camalytics/CamalyticsEnv/Projects/Sports/NFL/Data/Teams/"

for year in nfl_years:
    for team in nfl_afc_abb:
        url = f"https://www.pro-football-reference.com/teams/{team}/{year}.htm"
        data = requests.get(url)

        file_path = nfl_team_folder + f"{team}-{year}" + ".html"

        with open(file_path, "w+", encoding="utf-8") as f:
            f.write(data.text)
        time.sleep(random.uniform(5, 7))
```
